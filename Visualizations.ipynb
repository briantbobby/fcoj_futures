```python
import matplotlib.pyplot as plt
import requests
import pandas as pd
import io
import datetime
import numpy as np
import matplotlib.dates as mdates
```

```python
r = requests.get('https://www.quandl.com/api/v3/datasets/CHRIS/ICE_OJ1.csv?api_key=9BZrqXiy4Mck8_br9oG9')
OJ1 = pd.read_csv(io.StringIO(r.text))
OJ1['Date'] = pd.to_datetime(OJ1['Date'],format='%Y-%m-%d')
OJ1 = OJ1.set_index('Date')
```

```python
r = requests.get('https://www.quandl.com/api/v3/datasets/CHRIS/ICE_OJ2.csv?api_key=9BZrqXiy4Mck8_br9oG9')
OJ2 = pd.read_csv(io.StringIO(r.text))
OJ2['Date'] = pd.to_datetime(OJ2['Date'],format='%Y-%m-%d')
OJ2 = OJ2.set_index('Date')
```

```python
r = requests.get('https://www.quandl.com/api/v3/datasets/CHRIS/ICE_OJ3.csv?api_key=9BZrqXiy4Mck8_br9oG9')
OJ3 = pd.read_csv(io.StringIO(r.text))
OJ3['Date'] = pd.to_datetime(OJ3['Date'],format='%Y-%m-%d')
OJ3 = OJ3.set_index('Date')
```

```python
r = requests.get('https://www.quandl.com/api/v3/datasets/CHRIS/ICE_OJ4.csv?api_key=9BZrqXiy4Mck8_br9oG9')
OJ4 = pd.read_csv(io.StringIO(r.text))
OJ4['Date'] = pd.to_datetime(OJ4['Date'],format='%Y-%m-%d')
OJ4 = OJ4.set_index('Date')
```

```python
OJ1vals = OJ1['2010-02':'2009-11'].sort_values(by='Date')
OJ2vals = OJ2['2010-02':'2009-11'].sort_values(by='Date')
OJ3vals = OJ3['2010-02':'2009-11'].sort_values(by='Date')
OJ4vals = OJ4['2010-02':'2009-11'].sort_values(by='Date')

fig = plt.figure(figsize=(18,5))
ax = fig.add_subplot(1,1,1)
plt.plot(OJ1vals.index,OJ1vals['Settle'],label='OJ1')
plt.plot(OJ2vals.index,OJ2vals['Settle'],label='OJ2')
plt.plot(OJ3vals.index,OJ3vals['Settle'],label='OJ3')
plt.plot(OJ4vals.index,OJ4vals['Settle'],label='OJ4')
plt.axvspan('2010-01-04','2010-01-09',color='c',label='lows below freeze')
plt.axvspan('2010-01-10','2010-01-14',color='c')
ax.xaxis.set_major_locator(mdates.DayLocator(interval = 15))
plt.title('Winter 2009-2010')
plt.legend()
```
